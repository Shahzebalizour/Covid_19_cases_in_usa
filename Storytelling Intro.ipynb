#  Understanding COVID-19 New Deaths in the USA



##  What's Going On
You probably heard about COVID-19the virus that's been affecting people worldwide. Well, it's also been impacting the United States a lot. We want to look at the number of new deaths from COVID-19 each month in the USA.

##  What We're Looking At
We're focusing on making this easy to understand for everyone. Whether you're a doctor, a decision-maker, a researcher, or just someone interested in what's happeningyou can follow along.

##  Where We Get Our Data
We got our information from a file called 'usacovid.csv.' It tells us how many new COVID-19 deaths happened each month.





## Explanation of 'usacovid.csv' Dataset

1. **Month**: This column represents the month number, indicating the chronological order of the data from January (Month 1) to December (Month 12).

2. **New_deaths**: In the 'New_deaths' column, you'll find the number of new COVID-19 deaths reported for each respective month. This value signifies how many people succumbed to the virus during that specific month.

3. **Cumulative New Deaths**: The 'Cumulative New Deaths' column provides a cumulative count of new COVID-19 deaths. It showcases the total number of new deaths starting from the beginning of the year (January) and accumulating as new months are included in the dataset.

This dataset is valuable for tracking the progression of COVID-19 over time.

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import dataset
dataset = pd.read_csv("usacovid.csv")
dataset
```

```python
"""We have a dataset about deaths of peoples in usa during covid. We will explore it by Story telling graph, Let's Explore it
   We will break the months by quarterly SO for that we can have good insights to understand the cases"""

# For making continueous graph we are starting from -1 
qua1 = dataset[0:3]
qua2 = dataset[2:6]
qua3 = dataset[5:9]
qua4 = dataset[8:12]
```

```python
# Create a graph
fiq, (ax1,ax2,ax3,ax4) = plt.subplots(4,1, figsize = (9,9))

# Create a graph of complete data in all graphs
axes = [ax1,ax2,ax3,ax4]
for ax in axes:
    ax.plot(dataset["Month"], dataset["New_deaths"], color = "#aafada")
    
    # Removing Labels / Numbers
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    
    # Removing Sticks/Params On border Lines
    ax.tick_params(left = 0, bottom = 0)
    
    # Remove Borders
    for loc in ["left", "right", "top", "bottom"]:
        ax.spines[loc].set_visible(False)

# Now Fill the  Data by quarterly in every graph
ax1.plot(qua1["Month"], qua1["New_deaths"], color = "#32fa05", linewidth = 2.8)
ax2.plot(qua2["Month"], qua2["New_deaths"], color = "#32fa05", linewidth = 2.8)
ax3.plot(qua3["Month"], qua3["New_deaths"], color = "#32fa05", linewidth = 2.8)
ax4.plot(qua4["Month"], qua4["New_deaths"], color = "#32fa05", linewidth = 2.8)

# Now put the information in graph so we can understand it easily
ax1.text(0.7, -72, "0", alpha = 0.7)
ax1.text(2.7, 145, "77", alpha = 0.7)

ax2.text(2.7, 145, "77", alpha = 0.7)
ax2.text(5.9, 920, "785", alpha = 0.7)

ax3.text(5.9, 920, "785", alpha = 0.7)
ax3.text(8.8, 880, "739", alpha = 0.7)

ax4.text(8.8, 880, "739", alpha = 0.7)
ax4.text(11.4, 2340, "2247", alpha = 0.7)

# Now Label months on data
ax1.text(1.2, -300, "Jan-March", rotation = 3, fontdict = {"weight" : "bold"})
ax2.text(3.3, 250, "Apr-Jun", rotation = 65, fontdict = {"weight" : "bold"})
ax3.text(7, 1000, "Jul-Sept", rotation = 9, fontdict = {"weight" : "bold"})
ax4.text(10, 1000, "Oct-Dec", rotation = 30, fontdict = {"weight" : "bold"})

# Here put the information of cummulative deaths by covid
cumu_cases = [77, 4154, 6682, 10905]
n = 0
for x in axes:
    x.text(5, 1900, "Cummulative Cases:" + str(cumu_cases[n]),\
        fontdict = {"family" : 'monospace', 'color' : '#406160', 'size' : 11, 'weight' : 'bold'})
    n += 1
    
# Set the title 
ax1.text(0, 3000, "The Virus Kills 908 People Each Month", fontdict = {'family' : 'serif', 'size' : 17, 'weight' : 'bold'})

# Sub Title
ax1.text(0, 2640, "New Deaths in US", fontdict = {"family" : 'serif', 'size' : 12, 'weight' : 'bold'})

plt.show()
```
